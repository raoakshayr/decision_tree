```python
#let us find the number of samples belonging to benign and the number belonging to malignant
n_data=len(data)

n_benign=sum(data['2.1']==2)
n_malignant=sum(data['2.1']==4)
print('The number of benign samples : ',n_benign,'The number of malignant samples : ',n_malignant)

#the two classes are not equally represented in the data
```

```python
#now let us make a train test split making sure to maintain class ratios in train and test

train_ratio=2/3
test_ratio=1/3

n_train_b=train_ratio*n_benign
n_test_b=train_ratio*n_benign

n_train_m=train_ratio*n_malignant
n_test_m=train_ratio*n_malignant

train=pd.DataFrame()
test=pd.DataFrame()

benign_count=0
malig_count=0

#let us keep track of indices for splitting 
for i in range (n_data):
    if(data['2.1'][i]==2):
        if (benign_count<n_train_b):
            train=train.append(data.loc[i])
            benign_count+=1
        else: test=test.append(data.loc[i])
    elif(data['2.1'][i]==4):
        if(malig_count<n_train_m):
            train=train.append(data.loc[i])
            malig_count+=1
        else: test=test.append(data.loc[i])
```

```python
# now we must implement a decision tree on the train data
class node:
    def __init__(self,attribute=None,threshold=None,label=None,left=None,right=None,root=False):
        self.attribute=attribute
        self.threshold=threshold
        self.label=label
        self.left=left
        self.right=right
        self.root=root
```

```python
def node_entropy1(m_node,X_train,root_node):
    nm=0
    nb=0
    for i in range(len(X_train)):
        p=root_node
        #go till leaf
        while (p.left!=None or p.right!=None):
            if (X_train.iloc[i][p.attribute]>p.threshold):
                p=p.right
            else:
                p=p.left
        if p==m_node:
            if X_train.iloc[i]['2.1']==2:
                 nb+=1
            else:
                nm+=1    
    if nb!=0 and nm!=0:
        ent=-(nb/(nb+nm)*np.log(nb/(nb+nm))+nm/(nb+nm)*np.log(nm/(nb+nm)))
    else:
        ent=0
    if(nb<nm):
        m_node.label='Malignant'
    else:
        m_node.label='Benign'
    print('label : ',m_node.label,'n_b : ',nb,'n_m',nm,'entropy : ',ent)
    return ent,nb+nm
```

```python
def node_entropy2(m_node,X_train,root_node):
    nm=0
    nb=0
    for i in range(len(X_train)):
        p=root_node
        #go till leaf
        while (p.left!=None or p.right!=None):
            if (X_train.iloc[i][p.attribute]>p.threshold):
                p=p.right
            else:
                p=p.left
        if p==m_node:
            if X_train.iloc[i]['2.1']==2:
                 nb+=1
            else:
                nm+=1    
    if nb!=0 and nm!=0:
        ent=(2*nb/(nb+nm)*(1-nb/(nb+nm)))
    else:
        ent=0
    if(nb<nm):
        m_node.label='Malignant'
    else:
        m_node.label='Benign'
    print('label : ',m_node.label,'n_b : ',nb,'n_m',nm,'entropy : ',ent)
    return ent,nb+nm
```

```python
def split_entropy1(s_node,X_train,root_node,attribute,threshold):
    nm1=0
    nb1=0
    nm2=0
    nb2=0
    #pdb.set_trace()
    for i in range(len(X_train)):
        p=root_node
        #go till leaf
        while (p.left!=None or p.right!=None):
            if (X_train.iloc[i][p.attribute]>p.threshold):
                p=p.right
            else:
                p=p.left
        if p==s_node:
            if X_train.iloc[i][attribute]>threshold:
                if X_train.iloc[i]['2.1']==2:
                     nb1+=1
                else:
                    nm1+=1
                p=p.right
            else:
                if X_train.iloc[i]['2.1']==2:
                       nb2+=1
                else:
                    nm2+=1
                p=p.left
    #pdb.set_trace()
    n_branch1=nb1+nm1
    n_branch2=nb2+nm2
    if n_branch1!=0 and n_branch2!=0:
        n_node=n_branch1+n_branch2
        s_ent1=-(n_branch1/n_node)*(nb1/n_branch1*np.log(nb1/n_branch1)+nm1/n_branch1*np.log(nm1/n_branch1))  
        s_ent2=-(n_branch2/n_node)*(nb2/n_branch2*np.log(nb2/n_branch2)+nm2/n_branch2*np.log(nm2/n_branch2))
        s_ent=s_ent1+s_ent2
    else:
        s_ent=1000
    return s_ent   
    
```

```python
def split_entropy2(s_node,X_train,root_node,attribute,threshold):
    nm1=0
    nb1=0
    nm2=0
    nb2=0
    #pdb.set_trace()
    for i in range(len(X_train)):
        p=root_node
        #go till leaf
        while (p.left!=None or p.right!=None):
            if (X_train.iloc[i][p.attribute]>p.threshold):
                p=p.right
            else:
                p=p.left
        if p==s_node:
            if X_train.iloc[i][attribute]>threshold:
                if X_train.iloc[i]['2.1']==2:
                     nb1+=1
                else:
                    nm1+=1
                p=p.right
            else:
                if X_train.iloc[i]['2.1']==2:
                       nb2+=1
                else:
                    nm2+=1
                p=p.left
    #pdb.set_trace()
    n_branch1=nb1+nm1
    n_branch2=nb2+nm2
    if n_branch1!=0 and n_branch2!=0:
        n_node=n_branch1+n_branch2
        s_ent1=(n_branch1/n_node)*(2*nb1/n_branch1*(1-nb1/n_branch1))  
        s_ent2=(n_branch2/n_node)*(2*nb2/n_branch2*(1-nb2/n_branch2))
        s_ent=s_ent1+s_ent2
    else:
        s_ent=1000
    return s_ent   
    
```

```python
def acc(root_node,X_train):
    n_corr=0
    for i in range(len(X_train)):

        p=root_node
        #go till leaf
        while (p.left!=None or p.right!=None):
            if (X_train.iloc[i][p.attribute]>p.threshold):
                p=p.right
            else:
                p=p.left
        if X_train.iloc[i]['2.1']==2:
            if p.label=='Benign':
                n_corr+=1
        else:
            if p.label=='Malignant':
                n_corr+=1
    
    return n_corr/len(X_train)
```
